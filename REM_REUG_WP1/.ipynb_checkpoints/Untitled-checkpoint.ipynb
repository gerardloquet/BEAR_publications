```python
#now plot REUG 
fig, ax = plt.subplots(figsize = (15,7))

x_axis = np.arange(100,10000,43)
mean_ = REM_REUG.loc[:,'100':'9990'].values.mean(axis = 0)
std_ = REM_REUG.loc[:,'100':'9990'].values.std(axis = 0)
plt.plot(x_axis, mean_, color = 'darkblue')
plt.fill_between(x_axis,mean_-std_, mean_+ std_, color = 'lightblue',alpha = 0.7)
ax.set_xscale('log')
ax.set_xlim([1e02,1e04])
ax.set_xticks([ 1e02, 1e03, 2e03, 3e03, 4e03, 5e03,  1e04])
ax.set_xticklabels([100, 1000, 2000, 3000, 4000, 5000, 10000], fontsize = 14)
ax.grid(True)

plt.show()
```

```python
def seek_cluster_number(X, max_clust):
    from sklearn.cluster import KMeans
    SSE = []
    km = []
    cnt = 0
    for idx in range(1, max_clust):
        cnt = cnt + 1        
        print('Iteration = ',cnt)
        tmp = KMeans(n_clusters = idx, \
                    init = 'k-means++',\
                    n_init = 10,\
                    max_iter = 100,\
                    random_state = 0)
        tmp.fit(X)
        SSE.append(tmp.inertia_)
        km.append(tmp)
    plt.plot(range(1,max_clust), SSE, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('SSE')
    plt.show()
    return km
```

```python
km = seek_cluster_number(REM_REUG.loc[:,'100':'9990'].values,10)
```
